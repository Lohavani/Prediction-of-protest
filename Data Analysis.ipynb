### Analysis 1 - Distribution of event codes across USA

```python
# splitup of event codes in USA 
years = df['Year'].unique().tolist()
events = [10,11,12,13,14]
df_select = df.loc[df['Actor2CountryCode']=='USA']

c = df_select['Actor2CountryCode'].groupby([df['Year'], df['EventRootCode']]).value_counts()
colours = ['plum', 'darkviolet', 'mediumorchid', 'violet', 'thistle']
i = 0
cnt = []

for i in range (0,40,5): # as there are 5 event codes
    cnt.append(c.values[i:i+5].tolist())
 
print(c)
#print(cnt)

cols1 = ['Event Codes', 10, 11, 12, 13, 14]

for i in range(len(years)):
    #print(i," ", index)
    cnt[i].insert(0,years[i])
print(cnt)

df_plot = pd.DataFrame(cnt, columns = cols1)
print(df_plot)

df_plot.plot(x = 'Event Codes',
        kind = 'bar',
        title = "Distribution of events across past years in the USA",
        stacked = False,
        color = colours,
        figsize = (16, 7.5),
        legend = True,
        fontsize = 14,
        width = 0.7,
        edgecolor = 'black')
plt.legend(['10 - DEMAND', '11 - DISAPPROVE', '12 - REJECT', '13 - THREATEN', '14 - PROTEST'], fontsize = 14)
#plt.legend(fontsize = 14)
plt.xlabel("Year", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
```

### Analysis 2 - top 5 countries with highest number of protests

```python
#df.groupby('Actor2CountryCode').plot

df_plot2 = df.loc[df['EventRootCode']==14]
a = df_plot2['EventRootCode'].groupby([df_plot2['Year'], df_plot2['Actor2CountryCode']]).value_counts()

a
```

```python
# top 5 countries for every year

years = df['Year'].unique()
countries = []
cnt = []
i = 0
color = ['Maroon', 'teal', 'rebeccapurple', 'sienna', 'olive', 'crimson', 'Orangered', 'dimgrey']

for x in years :
    b = a[x].nlargest(5)
    cnt.append(b.values.tolist())
    temp = []
    var = b.index.values
    for i in range (len(b)):
        temp.append(var[i][0])
    countries.append(temp)
print (countries)
print (cnt)

fig, axs = plt.subplots(2, 4, figsize=(14, 6))
ind = np.arange(len(cnt))
width1 = 0.7
index = 0
for i in range (2):
    for j in range (4):
        axs[i][j].bar(countries[index], cnt[index], color = color[index], width = width1, edgecolor='black')
        axs[i][j].set_title(years[index])
        index = index + 1 
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.3)
plt.show()
```

### Analysis 3 - top 5 actors affected / involved in protests

```python
# actors who have been most affected in the USA 

df_plot2 = df_plot2.loc[df_plot2['Actor2CountryCode'] == 'USA']
years = df_plot2["Year"].unique().tolist()
#print(years)
#e = df_plot2['EventRootCode'].groupby([df_plot2['Year'], df_plot2['Actor2CountryCode']]).value_counts()
e = df_plot2['Actor1Name'].groupby(df_plot2['Year']).value_counts()
colors = ['pink', 'purple', 'silver', 'steelblue', 'khaki']

for yr in years :
    e[yr].nlargest(5).plot(kind='pie', figsize=(4,4), legend = True, colors = colors, labels = None, autopct='%1.0f%%')
    plt.title(yr)
    plt.legend(loc="upper right", bbox_to_anchor=(1.3,1), labels = e[yr].nlargest(5).index)
    plt.savefig("pie charts {}.png".format(yr))
    plt.show()
```
